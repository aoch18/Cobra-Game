```python
import pygame
from pygame.locals import *
from pygame.math import Vector2 as vetor
import random

class Cobra:
    def __init__(self):
        self.body = [vetor(100,100),vetor(100,110),vetor(100,120)]
        self.direction= vetor(0,-20)
        self.add = False
        
    def draw(self,janela,cor):
        for bloco in self.body:
            pygame.draw.rect(janela,cor,(bloco.x,bloco.y,20,20))
            
    def move(self):
        print(self.body[0])
        if self.add==True:
            estado_novo=self.body
            estado_novo.insert(0,estado_novo[0]+self.direction)
            self.body=estado_novo
            self.add=False
        else:
            estado_novo=self.body[:-1]
            estado_novo.insert(0,estado_novo[0]+self.direction)
            self.body=estado_novo
            #print(estado_novo)
        
    def move_up(self):
        self.direction= vetor(0,-20)
        
    def move_down(self):
        self.direction= vetor(0,20)
        
    def move_right(self):
        self.direction= vetor(20,0)
        
    def move_left(self):
        self.direction= vetor(-20,0)
        
    def die(self,base,altura):
        if self.body[0].x >=base or self.body[0].y>= altura or self.body[0].x < 0 or self.body[0].y <0:
            return True
        #cobra se pega ela mesma
        for i in self.body[1:]:
            if self.body[0]==i:
                return True
            
class Maça:
    def __init__(self,base,altura):
        self.generate(base,altura)
        
    def draw(self,janela,cor):
        pygame.draw.rect(janela,cor,(self.pos.x,self.pos.y,20,20))
        
    def generate(self,base,altura):
        self.x=random.randrange(0,base/20)
        self.y=random.randrange(0,altura/20)
        self.pos=vetor(self.x*20,self.y*20)
        print(self.pos,'maca')
        
    def check_colision(self,cobra,base,altura):
        if cobra.body[0]==self.pos:
            self.generate(base,altura)
            cobra.add=True
            return True
        
        for bloco in cobra.body[1:]:
            if self.pos==bloco:
                self.generate(base,altura)
        return False

pygame.init()

score_text = pygame.font.SysFont("Russo One",30)
game_over_text = pygame.font.SysFont("Russo One",25)

eat_sound = pygame.mixer.Sound("moeda.wav")
game_over_sound = pygame.mixer.Sound("game_over.wav")

score=0
base=720
altura=480
janela= pygame.display.set_mode((base,altura))
pygame.display.set_caption('Cobra')
cor=(0,255,0)
cor_maça=(255,0,0)
minha_cobra=Cobra()
minha_maça=Maça(base,altura)
tempo=pygame.time.Clock()
pause=False
game_over=False
fps=10
next_level=False

while True:
    tempo.tick(fps)
    
    if score%3==0 and score!=0 and next_level==True:
        fps+=1
        next_level=False
    elif game_over==True:
        fps=10
        
    for event in pygame.event.get():
        if event.type == QUIT:
            pygame.quit()
            
        if event.type == KEYDOWN and minha_cobra.direction.y != 20:
            if event.key == K_UP:
                minha_cobra.move_up()    
                
        if event.type == KEYDOWN and minha_cobra.direction.y != -20:
            if event.key == K_DOWN:
                minha_cobra.move_down()
                
        if event.type == KEYDOWN:
            if event.key == K_RIGHT and minha_cobra.direction.x != -20:
                minha_cobra.move_right()
                
        if event.type == KEYDOWN:
            if event.key == K_LEFT and minha_cobra.direction.x != 20:
                minha_cobra.move_left()
                
        if event.type == KEYDOWN:
            if event.key == K_p and pause==False:
                pause=True
            elif event.key == K_p and pause==True:
                pause=False
                
        if event.type == KEYDOWN:
            if event.key == K_s and game_over==True:
                minha_cobra=Cobra()
                score=0
                game_over=False
    if pause==False and game_over==False:
        
        janela.fill((52, 152, 219))
        minha_cobra.draw(janela,cor)
        minha_maça.draw(janela,cor_maça)
        minha_cobra.move()
    
        if minha_maça.check_colision(minha_cobra,base,altura):
            score+=1
            next_level=True
            eat_sound.play()
    
        if minha_cobra.die(base,altura):
            #pygame.quit()
            game_over=True
            game_over_sound.play()
        text = score_text.render("Score: {}".format(score),1,(255,255,255))
        janela.blit(text,(base-text.get_width()-20,20))
        
        text = score_text.render("Velocidad: {}".format(fps),1,(255,255,255))
        janela.blit(text,(base-text.get_width()-20,40))
    elif pause==True and game_over==False:
        text = score_text.render("PAUSED",1,(255,255,255))
        janela.blit(text,(base-text.get_width()-360,10))
    elif game_over==True:
        text = score_text.render("GAME OVER",1,(255,255,255))
        janela.blit(text,(base-text.get_width()-360,20))
        text = game_over_text.render("Play again",1,(255,255,255))
        janela.blit(text,(base-text.get_width()-380,40))
        text = game_over_text.render("Press S",1,(255,255,255))
        janela.blit(text,(base-text.get_width()-380,60))
        
    pygame.display.update()
```
